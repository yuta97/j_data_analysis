```python
import pandas as pd
# csvの読み取り
df= pd.read_csv("j_kawasaki2020.csv")
df.head()
```

```python
from datetime import date
from dateutil.relativedelta import relativedelta

import datetime

def create_age(birthday):
    
    today = datetime.date.today()
    birthday = datetime.datetime.strptime(birthday, '%Y/%m/%d')
    age = relativedelta(today,birthday)
    return age.years

df['age'] = df['birthday'].apply(create_age)
df.head()
```

```python
df['age'].mean()
```

```python
from datetime import date
from dateutil.relativedelta import relativedelta

import datetime

def create_age(birthday):
    
    today = datetime.date.today()
    birthday = datetime.datetime.strptime(birthday, '%Y/%m/%d')
    age = relativedelta(today,birthday)
    return age.years

df['age'] = df['birthday'].apply(create_age)
df.head()
```
